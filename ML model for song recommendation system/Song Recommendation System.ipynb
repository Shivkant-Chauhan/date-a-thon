```python
import numpy as np
import pandas as pd

data=pd.read_csv('/Users/soham2312/Downloads/tcc_ceds_music.csv',usecols=['artist_name','track_name','release_date','genre'])

data['tags']=data['artist_name']+","+data['genre']
data=data.drop(['artist_name','genre'],axis=1)

import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    t=[]
    for i in text.split():
        t.append(ps.stem(i))
    return " ".join(t)
data['tags']=data['tags'].apply(stem)
# data['track_name']=data['track_name'].apply(stem)

data['tags']=data['tags'].apply(lambda x:x.lower())
from sklearn.feature_extraction.text import CountVectorizer 
cv=CountVectorizer(max_features=15000,stop_words='english')
vectors=cv.fit_transform(data['tags']).toarray()

from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity

def recommend(song):
    song_index=data[data['track_name']==song].index[0]
    distances=similarity[song_index]
    song_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in song_list:
        print(data.iloc[i[0]].track_name)
    
recommend('payphone')
import pickle
pickle.dump(data.to_dict(),open('songs_dict.pkl','wb'))
pickle.dump(similarity,open('similarity1.pkl','wb'))
data
```
